# DATA PREPROCESSING

```python
# merging both the dataframes on the basis of comman column 'title'
movies = movies.merge(credits, on='title')
```

```python
movies.head(1)
```

```python
# now i'll get only those columns that are most required for the CONTENT BASED REMMONDED SYSTEM
# genres
# id
# keywords
# title
# overview
# cast
# crew

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
# above found that there are 3 movies that doesn't contain an overview, so will drop them
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# ['Action', 'Adventure', 'Fantasy', 'SciFi']
```

```python
# the above is a string containing a dictionary within a list 
# so will convert the string into int to apply a loop in the list
# ast.literal_eval will be used for that
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
# now will use the same func created above CONVERT() for the keywords column 
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# for the cast column there are many names of the cast in the dictionary format
# so will pick the top three of them only
# will make another function 

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# now for the last column CREW
# will fetch only the name of the DIRECTOR
# will create another func for it

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# column OVERVIEW is in a string format
# so will convert it into a list so that all the columns can be concatenate later
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing spaces between names like
# Sam Worthinton and Sam Mendes in all columns
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
movies.head()
```

```python
# now concatinating all the columns 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# removing the rest columns
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
# converting the LIST of TAGS column in STRING
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
# converting the content of TAGS column in lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

# VECTORIZATION

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```
